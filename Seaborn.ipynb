# Distribution Plot

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde = True, bins = 50, hist_kws=dict(edgecolor="k", linewidth=1))
```

```python
sns.jointplot(x='total_bill' , y='tip',data =tips, kind = 'scatter' )
```

```python
sns.pairplot(tips, hue = 'sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

```python
tips.head()
```

```python
import numpy as np
sns.barplot(x='sex', y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data = tips)
```

```python
sns.violinplot(x= 'day', y='total_bill',data=tips, hue ='sex', split=True)
```

```python
sns.stripplot(x= 'day', y='total_bill', data= tips, jitter=True, hue = 'sex', dodge = True)
```

```python
sns.violinplot(x= 'day', y = 'total_bill', data = tips)
sns.swarmplot(x= 'day', y = 'total_bill', data = tips, color= "black")
```

# Matrix Plot

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
#We need the data to be in matrix form. We can use either correlation or pivot to generate it
tc = tips.corr()
```

```python
sns.heatmap(tc, annot= True, cmap = "coolwarm")
```

```python
flight.head()
```

```python
flight_mat = flight.pivot_table(index= 'month', columns='year', values = 'passengers')
```

```python
sns.heatmap(flight_mat, cmap = 'magma', linecolor='white', linewidths=1)
```

```python
sns.clustermap(flight_mat, cmap = 'coolwarm', standard_scale=1)
```

# Grid plot

```python
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
tips.head()
```

```python
f = sns.FacetGrid(data = tips, col = 'time', row= 'smoker')
f.map(plt.scatter, 'total_bill', 'tip')
```
